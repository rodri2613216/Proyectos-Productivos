## 1. Velocidad final de un auto

### Enunciado
Determina la velocidad final de un auto si su velocidad inicial es 0 km/h, su aceleración es 0.8 m/s² y el tiempo transcurrido es de 30 segundos.

### Análisis
**Entradas:** v0 (km/h), a (m/s²), t (s)

**Proceso:** Convertir v0 a m/s (v0/3.6) y aplicar MRUA: vf = v0 + a·t. Convertir el resultado a km/h.

**Salidas:** Velocidad final en m/s y en km/h.

### Seudocódigo
```
Inicio
   Leer v0, a, t
   v0_ms <- v0 / 3.6
   vf_ms <- v0_ms + a * t
   vf_kmh <- vf_ms * 3.6
   Escribir "Velocidad final:", vf_ms, "m/s", vf_kmh, "km/h"
Fin
```

### Diagrama de flujo
Ejecuta la celda siguiente para generarlo.

````python
import graphviz
fd = graphviz.Digraph()
fd.attr(rankdir="TB")
fd.node("ini", "Inicio", shape="oval")
fd.node("in", "Leer v0, a, t", shape="parallelogram")
fd.node("p1", "v0_ms = v0 / 3.6", shape="box")
fd.node("p2", "vf_ms = v0_ms + a*t", shape="box")
fd.node("p3", "vf_kmh = vf_ms * 3.6", shape="box")
fd.node("out", "Escribir vf_ms, vf_kmh", shape="parallelogram")
fd.node("fin", "Fin", shape="oval")
fd.edges([("ini","in"),("in","p1"),("p1","p2"),("p2","p3"),("p3","out"),("out","fin")])
fd
````

### Prueba de escritorio
Datos de prueba: v0 = 0, a = 0.8, t = 30

| Paso | Operación | v0_ms | vf_ms | vf_kmh |
|:----:|-----------|:-----:|:-----:|:------:|
| 1 | Leer v0=0, a=0.8, t=30 | – | – | – |
| 2 | v0_ms = 0/3.6 | 0 | – | – |
| 3 | vf_ms = 0 + 0.8·30 | 0 | 24 | – |
| 4 | vf_kmh = 24·3.6 | 0 | 24 | 86.4 |
| 5 | Escribir | – | **24 m/s** | **86.4 km/h** |

**Código Python (ejecutable):**

````python
# Ejercicio 1 - Velocidad final
v0 = float(input("Velocidad inicial (km/h): "))   # 0 segun el problema
a  = float(input("Aceleracion (m/s2): "))          # 0.8
t  = float(input("Tiempo (s): "))                  # 30

v0_ms = v0 / 3.6
vf_ms = v0_ms + a * t        # MRUA: vf = v0 + a*t
vf_kmh = vf_ms * 3.6

print(f"Velocidad final: {vf_ms:.2f} m/s  ({vf_kmh:.2f} km/h)")
````

## 5. Inversión en el banco (2% mensual, un año)

### Enunciado
Un individuo invierte su capital en un banco que paga 2% mensual. Determinar cuánto dinero ganará después de un año.

### Análisis
**Entradas:** capital (monto invertido)

**Proceso:** Aplicar interés compuesto mensual durante 12 meses: monto = capital·(1+0.02)¹². La ganancia es monto − capital. (Si fuese interés simple: capital·0.02·12.)

**Salidas:** Monto final tras 12 meses y ganancia obtenida.

### Seudocódigo
```
Inicio
   Leer capital
   tasa <- 0.02
   monto_final <- capital * (1 + tasa) ^ 12
   ganancia <- monto_final - capital
   Escribir "Monto final:", monto_final, "Ganancia:", ganancia
Fin
```

### Diagrama de flujo
Ejecuta la celda siguiente para generarlo.

````python
import graphviz
fd = graphviz.Digraph()
fd.attr(rankdir="TB")
fd.node("ini", "Inicio", shape="oval")
fd.node("in", "Leer capital", shape="parallelogram")
fd.node("p1", "monto_final = capital*(1.02)^12", shape="box")
fd.node("p2", "ganancia = monto_final - capital", shape="box")
fd.node("out", "Escribir monto_final, ganancia", shape="parallelogram")
fd.node("fin", "Fin", shape="oval")
fd.edges([("ini","in"),("in","p1"),("p1","p2"),("p2","out"),("out","fin")])
fd
````

### Prueba de escritorio
Datos de prueba: capital = 1000 (interés compuesto)

| Variable | Operación | Valor |
|----------|-----------|:-----:|
| monto_final | 1000·(1.02)¹² | 1268.24 |
| ganancia | 1268.24 − 1000 | **268.24** |

*(Con interés simple sería 1000·0.02·12 = 240 de ganancia.)*

**Código Python (ejecutable):**

````python
# Ejercicio 5 - Inversion 2% mensual durante un año (interes compuesto)
capital = float(input("Capital a invertir: "))

tasa = 0.02
meses = 12
monto_final = capital * (1 + tasa) ** meses
ganancia = monto_final - capital

print(f"Monto final tras 12 meses: S/{monto_final:.2f}")
print(f"Ganancia: S/{ganancia:.2f}")
````

## 8. Área del terreno (Forma A)

### Enunciado
Una constructora vende terrenos con la Forma A (trapecio rectángulo) de la figura. Obtener el área ingresando los 3 lados A, B y C. El terreno equivale a un rectángulo de altura C más un triángulo de altura (A−C), ambos de base B.

### Análisis
**Entradas:** A (lado vertical izquierdo), B (base), C (lado vertical derecho)

**Proceso:** Aplicar el área del trapecio rectángulo: área = (A + C)/2 · B (equivale a rectángulo C·B más triángulo (A−C)·B/2).

**Salidas:** Área del terreno.

### Seudocódigo
```
Inicio
   Leer A, B, C
   area <- (A + C) / 2 * B
   Escribir "Area:", area
Fin
```

Ejecuta la celda siguiente para generarlo.

````python
import graphviz
fd = graphviz.Digraph()
fd.attr(rankdir="TB")
fd.node("ini", "Inicio", shape="oval")
fd.node("in", "Leer A, B, C", shape="parallelogram")
fd.node("p1", "area = (A + C)/2 * B", shape="box")
fd.node("out", "Escribir area", shape="parallelogram")
fd.node("fin", "Fin", shape="oval")
fd.edges([("ini","in"),("in","p1"),("p1","out"),("out","fin")])
fd
````

### Prueba de escritorio

| Variable | Operación | Valor |
|----------|-----------|:-----:|
| area | (10+6)/2 · 8 = 8·8 | **64** |

*Comprobación:* rectángulo 6·8=48 + triángulo (10−6)·8/2=16 → 64 ✔

## 10. Denominaciones del cajero automático

### Enunciado
Un cajero debe entregar una cantidad de dinero usando la menor cantidad de billetes/monedas posible, con denominaciones 200, 100, 50, 20, 10, 5 y 1. El usuario ingresa la cantidad a retirar.

### Análisis
**Entradas:** cantidad (monto a retirar, entero)

**Proceso:** Estrategia voraz (greedy): recorrer las denominaciones de mayor a menor; para cada una, calcular cuántas caben (división entera) y quedarse con el residuo para la siguiente.

**Salidas:** Cantidad de piezas por cada denominación.

### Seudocódigo
```
Inicio
   Leer cantidad
   restante <- cantidad
   denominaciones <- [200, 100, 50, 20, 10, 5, 1]
   Para cada d en denominaciones hacer
       n <- restante div d
       restante <- restante mod d
       Escribir "Piezas de", d, ":", n
   Fin Para
Fin
```

### Diagrama de flujo
Ejecuta la celda siguiente para generarlo.

````python
import graphviz
fd = graphviz.Digraph()
fd.attr(rankdir="TB")
fd.node("ini", "Inicio", shape="oval")
fd.node("in", "Leer cantidad", shape="parallelogram")
fd.node("p0", "restante = cantidad\ndenoms = [200,100,50,20,10,5,1]", shape="box")
fd.node("dec", "¿quedan\ndenominaciones?", shape="diamond")
fd.node("p1", "d = siguiente denominacion\nn = restante // d\nrestante = restante % d", shape="box")
fd.node("out", "Escribir d, n", shape="parallelogram")
fd.node("fin", "Fin", shape="oval")
fd.edge("ini","in"); fd.edge("in","p0"); fd.edge("p0","dec")
fd.edge("dec","p1", label="Si"); fd.edge("p1","out"); fd.edge("out","dec")
fd.edge("dec","fin", label="No")
fd
````

### Prueba de escritorio
Datos de prueba: cantidad = 388

| Denominación | restante antes | n = restante // d | restante después |
|:------------:|:--------------:|:-----------------:|:----------------:|
| 200 | 388 | 1 | 188 |
| 100 | 188 | 1 | 88 |
| 50  | 88  | 1 | 38 |
| 20  | 38  | 1 | 18 |
| 10  | 18  | 1 | 8 |
| 5   | 8   | 1 | 3 |
| 1   | 3   | 3 | 0 |

Resultado: 1×200, 1×100, 1×50, 1×20, 1×10, 1×5 y 3×1.

**Código Python (ejecutable):**

````python
# Ejercicio 10 - Cajero automatico (menor numero de piezas)
cantidad = int(input("Cantidad a retirar: S/"))

denominaciones = [200, 100, 50, 20, 10, 5, 1]
restante = cantidad
for d in denominaciones:
    n = restante // d        # division entera
    restante = restante % d  # residuo para la siguiente denominacion
    print(f"Piezas de {d}: {n}")
````
