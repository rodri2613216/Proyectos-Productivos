## 1. Llantera
Precio por llanta: **S/150** si se compran menos de 5; **S/120** si se compran 5 o más.

```python
# Ejercicio 1 - Llantera
cantidad = int(input("Cantidad de llantas a comprar: "))

if cantidad < 5:
    precio = 150
else:
    precio = 120

total = cantidad * precio
print(f"Precio por llanta: S/{precio:.2f}")
print(f"TOTAL A PAGAR: S/{total:.2f}")
```

## 2. Mayor y menor de 3 números

```python
# Ejercicio 2 - Mayor y menor de tres numeros
a = float(input("Numero 1: "))
b = float(input("Numero 2: "))
c = float(input("Numero 3: "))

# Determinamos el mayor con condicionales
if a >= b and a >= c:
    mayor = a
elif b >= a and b >= c:
    mayor = b
else:
    mayor = c

# Determinamos el menor con condicionales
if a <= b and a <= c:
    menor = a
elif b <= a and b <= c:
    menor = b
else:
    menor = c

print(f"El mayor es: {mayor}")
print(f"El menor es: {menor}")
```

## 3. Tributación
Tributa quien sea **mayor de 18 años** y tenga ingresos **≥ S/3000** mensuales.
El monto a tributar es siempre el **5%** del ingreso mensual.

```python
# Ejercicio 3 - Tributacion
edad = int(input("Edad: "))
ingresos = float(input("Ingresos mensuales (S/): "))

# "Mayor de 18" se interpreta como estrictamente mayor (> 18)
if edad > 18 and ingresos >= 3000:
    monto = ingresos * 0.05
    print("El usuario SI debe tributar.")
    print(f"Monto a tributar (5%): S/{monto:.2f}")
else:
    print("El usuario NO debe tributar.")
```

## 4. Colegiatura
Cada materia cuesta **S/100**. Si el promedio del alumno es **≥ 18**, recibe un
descuento del **50% sobre la mitad de los cursos**; si es **< 18**, paga la colegiatura completa.

```python
# Ejercicio 4 - Colegiatura
materias = int(input("Numero de materias que cursa: "))
promedio = float(input("Promedio del ultimo periodo: "))

costo_total = materias * 100  # costo sin descuento

if promedio >= 18:
    # 50% de descuento aplicado solo a la mitad de los cursos
    descuento = (materias / 2 * 100) * 0.50
    pago = costo_total - descuento
    print(f"Colegiatura completa: S/{costo_total:.2f}")
    print(f"Descuento aplicado:  S/{descuento:.2f}")
else:
    pago = costo_total
    print(f"Colegiatura completa: S/{costo_total:.2f}")
    print("Sin descuento (promedio menor a 18).")

print(f"TOTAL A PAGAR: S/{pago:.2f}")
```

## 5. Clasificación de triángulo
A partir de los 3 lados:
- **Equilátero** → mostrar el perímetro
- **Isósceles** → indicar qué lados son iguales
- **Escaleno** → mostrar el área (fórmula de Herón)

Se valida primero que los lados formen un triángulo válido.

```python
# Ejercicio 5 - Clasificacion de triangulo
import math

a = float(input("Lado a: "))
b = float(input("Lado b: "))
c = float(input("Lado c: "))

# Desigualdad triangular: cada lado debe ser menor que la suma de los otros dos
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        perimetro = a + b + c
        print("Triangulo EQUILATERO")
        print(f"Perimetro: {perimetro}")
    elif a == b or b == c or a == c:
        print("Triangulo ISOSCELES")
        if a == b:
            print("Los lados iguales son: a y b")
        elif b == c:
            print("Los lados iguales son: b y c")
        else:
            print("Los lados iguales son: a y c")
    else:
        s = (a + b + c) / 2  # semiperimetro
        area = math.sqrt(s * (s - a) * (s - b) * (s - c))
        print("Triangulo ESCALENO")
        print(f"Area: {area:.2f}")
else:
    print("Los lados ingresados NO forman un triangulo valido.")
```

## 6. Hora N segundos después
Se ingresa una hora (HH MM SS) y un número N de segundos; se muestra la hora resultante
(usando aritmética modular de 24 horas).

```python
# Ejercicio 6 - Hora N segundos despues
h = int(input("Horas (HH): "))
m = int(input("Minutos (MM): "))
s = int(input("Segundos (SS): "))
n = int(input("Segundos a sumar (N): "))

# Pasamos todo a segundos, sumamos y usamos modulo 86400 (24 horas)
total_seg = (h * 3600 + m * 60 + s + n) % 86400

nh = total_seg // 3600
nm = (total_seg % 3600) // 60
ns = total_seg % 60

print(f"Hora resultante: {nh:02d}:{nm:02d}:{ns:02d}")
```

## 7. Pago de una llamada
Tarifas por minuto:
- Minutos **1–5**: S/1.00 c/u
- Minutos **6–8** (siguientes 3): S/0.80 c/u
- Minutos **9–10** (siguientes 2): S/0.70 c/u
- Minuto **11 en adelante**: S/0.50 c/u

Impuesto: **3%** si es domingo y **5%** si es domingo de turno nocturno.

```python
# Ejercicio 7 - Pago de una llamada
minutos = int(input("Duracion de la llamada (minutos enteros): "))

# Cobro escalonado por tramos
if minutos <= 5:
    costo = minutos * 1.00
elif minutos <= 8:
    costo = 5 * 1.00 + (minutos - 5) * 0.80
elif minutos <= 10:
    costo = 5 * 1.00 + 3 * 0.80 + (minutos - 8) * 0.70
else:
    costo = 5 * 1.00 + 3 * 0.80 + 2 * 0.70 + (minutos - 10) * 0.50

# Impuesto segun el dia
impuesto = 0.0
domingo = input("Es domingo? (s/n): ").strip().lower()
if domingo == "s":
    nocturno = input("Es turno nocturno? (s/n): ").strip().lower()
    if nocturno == "s":
        impuesto = costo * 0.05   # domingo nocturno
    else:
        impuesto = costo * 0.03   # domingo normal

total = costo + impuesto
print(f"Costo por consumo: S/{costo:.2f}")
print(f"Impuesto:          S/{impuesto:.2f}")
print(f"TOTAL A PAGAR:     S/{total:.2f}")
```

## 8. Ordenar 3 números enteros diferentes
Se validan 3 enteros distintos y se muestran ordenados de forma **ascendente** o
**descendente** según indique el usuario.

```python
# Ejercicio 8 - Ordenar tres enteros diferentes
a = int(input("Numero entero 1: "))
b = int(input("Numero entero 2: "))
c = int(input("Numero entero 3: "))

if a == b or b == c or a == c:
    print("ERROR: los tres numeros deben ser diferentes entre si.")
else:
    orden = input("Orden deseado - ascendente (a) o descendente (d): ").strip().lower()
    numeros = [a, b, c]

    if orden == "a":
        numeros.sort()                 # ascendente
        print("Orden ascendente:", numeros)
    elif orden == "d":
        numeros.sort(reverse=True)     # descendente
        print("Orden descendente:", numeros)
    else:
        print("Opcion de orden no valida (use 'a' o 'd').")
```

## 9. Compra de N camisas con descuento
Descuento según cantidad:
- **1 a 4** camisas → 12.5%
- **5 a 8** camisas → 20%
- **más de 8** camisas → 31.5%

> El enunciado no especifica el precio unitario, por lo que se solicita al usuario.

```python
# Ejercicio 9 - Compra de N camisas con descuento
n = int(input("Numero de camisas a comprar: "))
precio = float(input("Precio por camisa (S/): "))

subtotal = n * precio  # compra sin descuento

if 1 <= n <= 4:
    porcentaje = 0.125
elif 5 <= n <= 8:
    porcentaje = 0.20
else:  # n > 8
    porcentaje = 0.315

monto_descuento = subtotal * porcentaje
total = subtotal - monto_descuento

print(f"Compra sin descuento: S/{subtotal:.2f}")
print(f"Descuento ({porcentaje*100:.1f}%): S/{monto_descuento:.2f}")
print(f"Compra con descuento: S/{total:.2f}")
```

## 10. Llamadas internacionales por zona
Según la **clave** de la zona geográfica y los minutos hablados, se muestra el nombre
de la zona y el costo total.

| Código | Zona geográfica   | Costo x minuto (S/) |
|:------:|-------------------|:-------------------:|
| 10     | América del Norte | 2.2 |
| 12     | América del Centro| 2.5 |
| 20     | América del Sur   | 1.2 |
| 22     | Asia              | 3.5 |
| 30     | Europa            | 3.0 |
| 32     | África            | 3.2 |

> La tabla del enunciado venía con las columnas desalineadas; se reconstruyó como se muestra arriba.

```python
# Ejercicio 10 - Llamadas internacionales por zona
# Diccionario: clave -> (nombre de zona, costo por minuto)
zonas = {
    10: ("America del Norte", 2.2),
    12: ("America del Centro", 2.5),
    20: ("America del Sur", 1.2),
    22: ("Asia", 3.5),
    30: ("Europa", 3.0),
    32: ("Africa", 3.2),
}

clave = int(input("Clave de la zona geografica: "))
minutos = int(input("Minutos hablados: "))

if clave in zonas:
    nombre, costo_min = zonas[clave]
    total = minutos * costo_min
    print(f"Zona: {nombre}")
    print(f"Costo por minuto: S/{costo_min:.2f}")
    print(f"COSTO TOTAL DE LA LLAMADA: S/{total:.2f}")
else:
    print("Clave de zona no valida.")
```
