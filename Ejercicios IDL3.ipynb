## 1. Suma y promedio de pares e impares (1 a N)
Recorre del 1 al N y acumula por separado los pares y los impares.

```python
# Ejercicio 1 - Suma y promedio de pares e impares de 1 a N
N = int(input("Ingrese N: "))

suma_pares = cant_pares = 0
suma_impares = cant_impares = 0

for i in range(1, N + 1):
    if i % 2 == 0:
        suma_pares += i
        cant_pares += 1
    else:
        suma_impares += i
        cant_impares += 1

prom_pares = suma_pares / cant_pares if cant_pares > 0 else 0
prom_impares = suma_impares / cant_impares if cant_impares > 0 else 0

print(f"PARES   -> Suma: {suma_pares} | Promedio: {prom_pares:.2f}")
print(f"IMPARES -> Suma: {suma_impares} | Promedio: {prom_impares:.2f}")
```

## 2. Clasificar 10 números
Cuenta cuántos son pares, impares, positivos, neutros (cero) y negativos.
Un número se cuenta en dos grupos: por paridad y por signo.

```python
# Ejercicio 2 - Clasificar 10 numeros
pares = impares = 0
positivos = neutros = negativos = 0

for i in range(1, 11):
    n = int(input(f"Numero {i}: "))

    # Paridad
    if n % 2 == 0:
        pares += 1
    else:
        impares += 1

    # Signo
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        neutros += 1

print(f"Pares: {pares}")
print(f"Impares: {impares}")
print(f"Positivos: {positivos}")
print(f"Neutros (cero): {neutros}")
print(f"Negativos: {negativos}")
```

## 3. Porcentajes respecto a un número X
Se ingresan N números y un valor X. Cada condición se cuenta de forma independiente
y se reporta como porcentaje del total.

> **Nota:** las condiciones *menor / mayor / igual* ya cubren todos los casos posibles,
> por lo que "no cumple ninguna condición" normalmente resulta 0%. Se calcula de forma
> literal (un número que no satisfaga ninguna de las cuatro condiciones listadas).

```python
# Ejercicio 3 - Porcentajes respecto a X
N = int(input("Cuantos numeros ingresara (N): "))
X = float(input("Ingrese el numero X: "))

menores = mayores = iguales = entre = ninguna = 0

for i in range(1, N + 1):
    n = float(input(f"Numero {i}: "))

    cond_menor  = n < X
    cond_mayor  = n > X
    cond_igual  = n == X
    cond_entre  = 1 <= n <= X

    if cond_menor:  menores += 1
    if cond_mayor:  mayores += 1
    if cond_igual:  iguales += 1
    if cond_entre:  entre += 1
    if not (cond_menor or cond_mayor or cond_igual or cond_entre):
        ninguna += 1

print(f"Menores que X:        {menores/N*100:.2f}%")
print(f"Mayores que X:        {mayores/N*100:.2f}%")
print(f"Iguales a X:          {iguales/N*100:.2f}%")
print(f"Entre 1 y X:          {entre/N*100:.2f}%")
print(f"No cumple ninguna:    {ninguna/N*100:.2f}%")
```

## 4. Máximo y mínimo de N números (simultáneo)
Se inicializan máximo y mínimo con el primer número y se actualizan en cada lectura.

```python
# Ejercicio 4 - Maximo y minimo simultaneos
N = int(input("Cuantos numeros ingresara (N): "))

maximo = minimo = None

for i in range(1, N + 1):
    n = float(input(f"Numero {i}: "))
    if i == 1:
        maximo = minimo = n
    else:
        if n > maximo:
            maximo = n
        if n < minimo:
            minimo = n

print(f"Maximo: {maximo}")
print(f"Minimo: {minimo}")
```

## 5. Ahorro anual con depósitos variables
Cada mes se deposita una cantidad distinta. Se muestra el acumulado tras cada depósito
y, al final, el ahorro total de los 12 meses.

```python
# Ejercicio 5 - Ahorro anual
acumulado = 0

for mes in range(1, 13):
    deposito = float(input(f"Deposito del mes {mes}: S/"))
    acumulado += deposito
    print(f"  Ahorro acumulado al mes {mes}: S/{acumulado:.2f}")

print(f"\nAHORRO FINAL DEL AÑO (12 meses): S/{acumulado:.2f}")
```

## 6. Cronómetro (hh:mm:ss)
Cuenta el tiempo en formato horas:minutos:segundos, actualizándose en el mismo lugar.
Es **referencial**: usa `time.sleep(1)`, por lo que no es exacto a un cronómetro real.
Para no quedar en bucle infinito en Colab, se ejecuta por una duración indicada.

```python
# Ejercicio 6 - Cronometro referencial
import time
from IPython.display import clear_output

duracion = int(input("Cuantos segundos debe correr el cronometro: "))

for t in range(duracion + 1):
    h = t // 3600
    m = (t % 3600) // 60
    s = t % 60
    clear_output(wait=True)
    print(f"{h:02d}:{m:02d}:{s:02d}")
    time.sleep(1)

print("Cronometro detenido.")
```

## 7. Factorial de dos números y prefijo común
Calcula el factorial de cada número y muestra la expansión `1x2x3...`. Como ambos
productos comparten el mismo inicio, son iguales hasta el múltiplo `mínimo(a, b)`.

```python
# Ejercicio 7 - Factorial de dos numeros y prefijo comun
def expansion_y_factorial(n):
    factores = [str(i) for i in range(1, n + 1)]
    producto = 1
    for i in range(1, n + 1):
        producto *= i
    return "x".join(factores), producto

a = int(input("Primer numero: "))
b = int(input("Segundo numero: "))

exp_a, fact_a = expansion_y_factorial(a)
exp_b, fact_b = expansion_y_factorial(b)

print(f"{a}! = {exp_a} = {fact_a}")
print(f"{b}! = {exp_b} = {fact_b}")

comun = min(a, b)
print(f">> Son iguales hasta el multiplo {comun}")
```

## 8. Asistentes a una fiesta
Se registran personas (edad y sexo) una por una; tras cada registro se pregunta si se
desea ingresar otra. Calcula total, hombres/mujeres, promedio de edad por sexo y las
edades mínima y máxima.

```python
# Ejercicio 8 - Asistentes a una fiesta
total = 0
hombres = mujeres = 0
suma_h = suma_m = 0
edad_min = edad_max = None

continuar = "s"
while continuar == "s":
    edad = int(input("Edad de la persona: "))
    sexo = input("Sexo (M=masculino / F=femenino): ").strip().upper()

    total += 1
    if sexo == "M":
        hombres += 1
        suma_h += edad
    elif sexo == "F":
        mujeres += 1
        suma_m += edad

    if edad_min is None or edad < edad_min:
        edad_min = edad
    if edad_max is None or edad > edad_max:
        edad_max = edad

    continuar = input("Desea registrar a otra persona? (s/n): ").strip().lower()

prom_h = suma_h / hombres if hombres > 0 else 0
prom_m = suma_m / mujeres if mujeres > 0 else 0

print(f"\nTotal de asistentes: {total}")
print(f"Hombres: {hombres} | Mujeres: {mujeres}")
print(f"Promedio de edad (hombres): {prom_h:.2f}")
print(f"Promedio de edad (mujeres): {prom_m:.2f}")
print(f"Edad menor: {edad_min} | Edad mayor: {edad_max}")
```

## 9. Producto con arrays: 100 × 98 × 96 × ... 
Se construye un array con los números pares de 100 hasta 2 y se calcula su producto.

> El enunciado termina en "× 1"; como el paso es de 2 en 2 desde 100, la secuencia llega
> hasta 2, y multiplicar por 1 no altera el resultado.

```python
# Ejercicio 9 - Producto con arrays
numeros = list(range(100, 0, -2))  # [100, 98, 96, ..., 4, 2]

producto = 1
for n in numeros:
    producto *= n

print("Array:", numeros)
print(f"\nProducto total: {producto}")
```

## 10. Notas de N estudiantes (5 a 20)
Lee N notas validando que cada una esté entre 0 y 20. Nota mínima aprobatoria: **13**.
Calcula los 11 resultados solicitados (a–k).

```python
# Ejercicio 10 - Notas de N estudiantes
# Validar N entre 5 y 20
while True:
    N = int(input("Cantidad de estudiantes (5 a 20): "))
    if 5 <= N <= 20:
        break
    print("  N debe estar entre 5 y 20.")

# Leer y validar notas
notas = []
for i in range(1, N + 1):
    while True:
        nota = int(input(f"Nota del estudiante {i} (0-20): "))
        if 0 <= nota <= 20:
            notas.append(nota)
            break
        print("  La nota debe estar entre 0 y 20.")

mayor = max(notas)
menor = min(notas)
cant_mayor = notas.count(mayor)
cant_menor = notas.count(menor)
promedio = sum(notas) / N
pares = sum(1 for n in notas if n % 2 == 0)
impares = N - pares
aprobados = sum(1 for n in notas if n >= 13)
desaprobados = N - aprobados
porc_aprob = aprobados / N * 100
porc_desap = desaprobados / N * 100

print("\n----- RESULTADOS -----")
print(f"a) Mayor nota: {mayor}")
print(f"b) Menor nota: {menor}")
print(f"c) Alumnos con la mayor nota: {cant_mayor}")
print(f"d) Alumnos con la menor nota: {cant_menor}")
print(f"e) Promedio del salon: {promedio:.2f}")
print(f"f) Notas pares: {pares}")
print(f"g) Notas impares: {impares}")
print(f"h) Aprobados: {aprobados}")
print(f"i) Desaprobados: {desaprobados}")
print(f"j) Porcentaje aprobados: {porc_aprob:.2f}%")
print(f"k) Porcentaje desaprobados: {porc_desap:.2f}%")
```
